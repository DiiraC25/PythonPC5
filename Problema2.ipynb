```python
import pandas as pd

df_wine = pd.read_csv("./data/winemag-data-130k-v2.csv")
df_paises = pd.read_csv("./data/paises.csv")

# Renombrar columnas
df_wine = df_wine.rename(columns={'country': 'país', 'description': 'descripción', 'designation': 'denominación', 'points': 'puntuación'})

# Crear nuevas columnas
# 1. Convertir los precios a euros (suponiendo que los precios están en dólares)
df_wine['precio_eur'] = df_wine['price'] * 0.88  # Tasa de cambio: 1 dólar = 0.88 euros
# 2. Calcular el precio por punto
df_wine['precio_por_punto'] = df_wine['precio_eur'] / df_wine['puntuación']

# Fusionar los DataFrames utilizando la columna 'país'
df_wine = df_wine.merge(df_paises[['nombre', 'continente']], left_on='país', right_on='nombre', how='left')
df_wine.drop(columns=['nombre'], inplace=True)

df_wine.head()
```

**Reportes:**

```python
# 1. Promedio de puntuación por país:
promedio_puntuacion_pais = df_wine.groupby('país')['puntuación'].mean().sort_values(ascending=False)
print(promedio_puntuacion_pais)
```

```python
# 2. Cantidad de vinos por país y continente:
cantidad_vinos_por_pais_continente = df_wine.groupby(['país', 'continente']).size().reset_index(name='cantidad_vinos')
print(cantidad_vinos_por_pais_continente)
```

```python
# 3. Precio promedio de vinos por continente:
precio_promedio_por_continente = df_wine.groupby('continente')['precio_eur'].mean().sort_values(ascending=False)
print(precio_promedio_por_continente)
```

```python
# 4. Cantidad de vinos por rango de puntuación:
bins = [0, 85, 90, 95, 100]
labels = ['80-85', '86-90', '91-95', '96-100']
df_wine['rango_puntuacion'] = pd.cut(df_wine['puntuación'], bins=bins, labels=labels)
cantidad_vinos_por_rango_puntuacion = df_wine.groupby('rango_puntuacion').size().reset_index(name='cantidad_vinos')
print(cantidad_vinos_por_rango_puntuacion)
```

**Almacenar en excel:**

```python
precio_promedio_por_continente = df_wine.groupby('continente')['precio_eur'].mean().sort_values(ascending=False)

precio_promedio_por_continente_df = precio_promedio_por_continente.reset_index(name='precio_promedio')
precio_promedio_por_continente_df.to_excel('precio_promedio_por_continente.xlsx', index=False)
```
